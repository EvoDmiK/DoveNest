```python
import json
import os

from bs4 import BeautifulSoup as bs
import requests as req
```

```python
BASE_URL    = 'https://www.nintendo.co.kr'
SALE_URL    = f'www.store.nintendo.co.kr/games/sale'

RELEASE_STEAM    = 'https://store.steampowered.com/explore/upcoming/?l=koreana'
RELEASE_NINTENDO = f'{BASE_URL}/software/release/'
```

```python
def preprocessing(string, dtype = 'date'):
    
    string = string.replace('  ', '')
    string = string.replace('(', '').replace(')', '')
    string = string.replace('\r\n', '')
    
    return string


def get_crawl(URL):
    res = req.get(URL).text
    return bs(res, 'html.parser')
```

```python
soup_rel = get_crawl(RELEASE_NINTENDO)
releases = soup_rel.select('.release-soft')[0].select('li')
```

```python
for release in releases:
    thumbnail = release.select('img')[0]['src']
    thumbnail = f'{BASE_URL}{thumbnail}'
    
    title     = preprocessing(release.select('.device-switch')[0].text)
    rel_date  = preprocessing(release.select('.rel-date')[0].text)
    developer = preprocessing(release.select('.rel-maker')[0].text)
    
    rel_date  = rel_date if rel_date != '\n' else '미정'
    
    print(f'[title]      : {title}')
    print(f'[rel date]   : {rel_date}')
    print(f'[thumbnail]  : {thumbnail}')
    print(f'[develpoper] : {developer}\n')
```

```python
soup_rel = get_crawl(RELEASE_STEAM)
releases = soup_rel.select('div.home_tabs_content > div.tab_content')[0].select('a.tab_item')
```

```python
for release in releases:
    
    title     = release.select('div.tab_item_name')[0].text
    rel_date  = release.select('div.release_date')[0].text
    thumbnail = release.select('img.tab_item_cap_img')[0]['src']
    
    print(f'[title]      : {title}')
    print(f'[rel date]   : {rel_date}')
    print(f'[thumbnail]  : {thumbnail}\n')
```
