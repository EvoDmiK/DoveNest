```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
res    = req.get('https://steamspy.com/').text
soup   = bs(res, 'html.parser') 
```

```python
datas  = list(soup.select('table.table-striped > tbody > tr > td'))
datas  = [data for data in datas if 'tplaytime' not in str(data)]
datas  = [datas[idx + 1 : idx + 5] for idx in range(0, len(datas), 6)]
```

```python
for idx, data in enumerate(datas[:100], 1):
    
    info, date, price, score = data
    
    title      = info.text
    released   = date.text
    price      = price.text
    
    try: 
        appid     = info.select('a')[0]['href'].split('/')[2]
        thumbnail = info.select('a > img')[0]['src'] 
    except: print(idx, title)
    
    print(idx, appid, title, released, price, thumbnail)
```
