```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
import os
```

```python
ROOT_PATH = '/home/jovyan/project'
DATA_PATH = f'{ROOT_PATH}/TIL/AI_study/dataset/steam/clean_data'
```

```python
df = pd.read_csv(f'{DATA_PATH}/steam.csv')
df.head()
```

```python
columns = ['steamspy_tags', 'name']
print('is null in these columns?', df[columns].isnull().values.any())
df[columns].head()
```

```python
def get_features(data):
    
    features = [f'{tag} {name}' for tag, name in zip(data['steamspy_tags'], data['name'])]
    return features

df['features'] = get_features(df)
df['Game_no']  = range(0, df.shape[0])

df.head()
```

```python
cm = CountVectorizer().fit_transform(df['features'])
cm.shape
```

```python
cs = cosine_similarity(cm)
```
